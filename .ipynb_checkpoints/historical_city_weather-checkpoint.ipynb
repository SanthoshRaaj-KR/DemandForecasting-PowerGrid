```python
import requests

result = requests.get(location_url).json()
components = result['response']['results']['results'][0]
lat = components['latitude']
lng = components['longitude']
print(lat,lng)
```

Get required data (can add/remove params)

```python
import pandas as pd

url = "https://archive-api.open-meteo.com/v1/era5"

params = {
    "latitude": lat,
    "longitude": lng,
    "start_date": "2020-01-01",
    "end_date": "2020-01-05",
    "daily": (
        "temperature_2m_mean,"
        "temperature_2m_max,"
        "dewpoint_2m_mean,"
        "relative_humidity_2m_mean,"
        "wind_speed_10m_mean,"
        "wind_gusts_10m_max,"
        "precipitation_sum,"
        "shortwave_radiation_sum"
    ),
    "timezone": "Asia/Kolkata"
}

# Make request
response = requests.get(url, params=params)
data = response.json()

# Convert to DataFrame
df_weather = pd.DataFrame(data["daily"])

# Convert date column to datetime
df_weather["time"] = pd.to_datetime(df_weather["time"])

# Rename time -> Date to match your power dataset
df_weather.rename(columns={"time": "Date"}, inplace=True)

print(df_weather)
print(df_weather.info())
```
